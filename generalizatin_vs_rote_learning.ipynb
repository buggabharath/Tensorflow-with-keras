```python
import numpy as np
```

```python
x = np.linspace(-1, 1, 100)
```

```python
signal = 2 + x + 2 * x * x
noise = np.random.normal(0, 0.1, 100)
y = signal + noise
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(signal, 'b')
plt.plot(y, 'g')
plt.plot(noise, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['Without Noise', 'With Noise', 'Noise'], loc = 2)
```
